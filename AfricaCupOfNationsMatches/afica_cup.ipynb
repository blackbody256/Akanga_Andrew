```python
null_attendance_rows = df[df['Attendance'].isnull()]
null_attendance_rows.head()
```

```python
# Select rows where HomeTeamGoals are between 3 and 6 inclusive
high_scoring_home = df[(df['HomeTeamGoals'] >= 3) & (df['HomeTeamGoals'] <= 6)]

# Alternative method using the between() function
# high_scoring_home = new_df[new_df['HomeTeamGoals'].between(3, 6)]

# Display the result
print(f"Number of matches where home team scored 3-6 goals: {len(high_scoring_home)}")
high_scoring_home.sample(5)
```

```python
# Change the AwayTeamGoals in the 3rd row (index 2) to 10
df.at[2, 'AwayTeamGoals'] = 10

# Display the updated row to verify the change
print("Updated 3rd row:")
print(df.iloc[2])
```

```python
#Grouping rows using hometeam name then sort them by their goals
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])


print("DataFrame sorted by HomeTeam (ascending) and HomeTeamGoals (descending):")
sorted_df.head(10)  # Show first 10 rows of the sorted DataFrame
```
